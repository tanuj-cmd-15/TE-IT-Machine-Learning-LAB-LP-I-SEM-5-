## Decision Tree:

```python
import pandas as pd
```

```python
A=pd.read_csv('C:\\Users\\desktop\\Downloads\\archive\\Admission_Predict.csv')
```

```python
A
```

```python
A.loc[A["Chance of Admit "]>=0.5, "Chance of Admit "]=1
```

```python
A
```

```python
A.loc[A["Chance of Admit "]<0.5, "Chance of Admit "]=0
```

```python
A["Chance of Admit "] = A["Chance of Admit "].astype(int)
```

```python
A.head(20)
```

## Spliting Dataset

```python
from sklearn.model_selection import train_test_split
x=A.drop('Chance of Admit ',axis=1)
y=A['Chance of Admit ']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=101)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

## Decision tree model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(xtrain,ytrain)
y_pred = classifier.predict(xtest)
```

```python
y_pred
```

```python
df = pd.DataFrame({'acutal':ytest,'predicted':y_pred})
```

```python
df.head()
```

```python
classifier.classes_
```

```python
from sklearn.tree import plot_tree
features = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research']
features
```

```python
plot_tree(classifier,feature_names=features)
```

## accuracy measures

```python
from sklearn.tree import export_graphviz
export_graphviz(classifier,)
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(ytest,y_pred)
```

```python
print(classification_report(ytest,y_pred))
```

## Visualization

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,15))
plot_tree(classifier, 
          feature_names = features, 
          max_depth=9,
          class_names=['0','1'],
          filled = True, 
          rounded = True,
         fontsize=8)

plt.savefig('tree_visualization.png') 
```
